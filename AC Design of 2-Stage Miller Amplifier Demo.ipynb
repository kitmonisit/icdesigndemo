```python
%load_ext autoreload
%autoreload 2

import sympy as sp
import acdesign
```

```python
gm1, go1, CM, gm2, YL, s = sp.symbols(r"g_{m1} g_{o1} C_M g_{m2} Y_L s")
ro1, RL, GL, CL = sp.symbols(r"r_{o1} R_L G_L C_L")
w, f = sp.symbols(r'\omega f')

G_exact = sp.Matrix(
    [
        [0, 0, 0, 1],
        [-gm1, go1 + s * CM, -s * CM, 0],
        [0, gm2 - s * CM, YL + s * CM, 0],
        [1, 0, 0, 0],
    ]
)

G_approx = sp.Matrix(
    [
        [0, 0, 0, 1],
        [-gm1, go1 + s * gm2 * RL * CM, 0, 0],
        [0, gm2, YL, 0],
        [1, 0, 0, 0],
    ]
)

amp_exact = acdesign.Amplifier(G_exact)
fn_exact = amp_exact.voltage_gain()

amp_approx = acdesign.Amplifier(G_approx)
fn_approx = amp_approx.voltage_gain()
```
