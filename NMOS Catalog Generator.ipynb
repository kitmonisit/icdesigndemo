```python
%load_ext autoreload
%autoreload 2

from sizinglib import build_nmos, build_figure, plotter, set_limits
```

```python
TARGET_VSTAR = [0.10, 0.15, 0.30]
TARGET_GM = 1e-3
TARGET_VDS = 1.0
TARGET_LENGTH = 1e-6

ds = list(map(
    lambda vstar: build_nmos(target_vstar=vstar,
                         target_gm=TARGET_GM,
                         target_vds=TARGET_VDS,
                         target_length=TARGET_LENGTH,
                        ),
    TARGET_VSTAR,
))
axes = build_figure()
plotlimits = list(map(
    lambda x: plotter(*x, axes=axes),
    enumerate(ds)
))
set_limits(axes, plotlimits)
```
